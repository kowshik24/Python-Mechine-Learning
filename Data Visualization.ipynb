```python
from matplotlib import pyplot as plt
```

```python
import numpy as np
```

```python
x = np.arange(1,11)
x
```

```python
y1 = 2*x
y2 = 3*x
y1,y2
```

```python
plt.plot(x,y1,color='orange',linewidth=3,linestyle=':')
plt.plot(x,y2,color='blue',linewidth=3,linestyle=':')
plt.title("Line Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)
plt.show()
```

```python
#bar-plot
```

```python
student = {"Sam":30 , "Bob":50, "Anne":70}
```

```python
names = list(student.keys())
marks = list(student.values())
names,marks
```

```python
plt.bar(names,marks,color='yellow')
plt.title("Marks of Students")
plt.xlabel("Names of Students")
plt.ylabel("Marks of Students")
plt.grid(True)
plt.show()
```

```python
plt.barh(names,marks,color='yellow') # Difference is 'barh'
plt.title("Marks of Students")
plt.xlabel("Names of Students")
plt.ylabel("Marks of Students")
plt.grid(True)
plt.show()
```

```python
#Scatter-Plot
```

```python
x = [2,1,4,5,6,7]
y1 = [8,9,2,4,5,6]
y2 = [2,4,5,6,7,8]

plt.scatter(x,y1,color='red')
plt.scatter(x,y2,color='blue')
plt.grid(True)
plt.title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
```
