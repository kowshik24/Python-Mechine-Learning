```python
import pandas as pd
df = pd.read_csv("Uber Drives 2016.csv")
df.head()
```

# Grouping(mean,max,min)

SAC(Split Apply Combine)
It will split the comumn and apply its condition





```python
df.groupby("START*")["MILES*"].agg("mean")
```

```python
df.groupby("START*")["MILES*"].agg("max")
```

```python
df.groupby("START*")["MILES*"].agg("min")
```

```python
df.groupby("START*")["MILES*"].mean()
```

```python
df.groupby("START*").mean() #It will also work cause the csv file has only one numeric column but it there more then one numeric
#comuln it will not work
```

```python
kowshik = df.groupby(["START*", "STOP*"]).agg(["mean","sum"]).head(n=10) # nested column
kowshik
```

```python
df = df.iloc[:-1,:]
df.tail()
```

```python
df
```

```python
df.dtypes
```

```python
df["START_DATE*"] = pd.to_datetime(df["START_DATE*"])
df.dtypes
```

```python
Date = df.groupby("START*").agg({"MILES*":"mean", "START_DATE*":["min","max"]})
Date.reset_index(inplace=True)
Date.columns = ["START_CITY","AVG_CITY","EARLIST_CITY","RECENT_CITY"]
Date
```

```python
df.groupby("START*").agg({"MILES*":"mean", "START_DATE*":"max"}).head()
```

```python
df.groupby("START*")[["MILES*","START_DATE*"]].agg(["mean","max"])
```
